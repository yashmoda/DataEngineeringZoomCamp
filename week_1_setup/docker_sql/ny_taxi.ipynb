```python
import pandas as pd
import sqlalchemy
```

```python
db_engine = sqlalchemy.create_engine('postgresql://root:root@localhost:5432/ny_taxi')
connection = db_engine.connect()
```

```python
trip_df = pd.read_csv('green_tripdata_2019-01.csv')
trip_df.shape
```

```python
trip_df['lpep_pickup_datetime'] = pd.to_datetime(trip_df['lpep_pickup_datetime'])
trip_df['lpep_dropoff_date'] = pd.to_datetime(trip_df['lpep_dropoff_datetime'])
trip_df.head()
```

```python
print(pd.io.sql.get_schema(trip_df, 'green_taxi_data', con=db_engine))
```

```python
trip_df.to_sql(name='green_taxi_data', con=db_engine, if_exists='replace')
```
